## Petit retour sur SQL et ses évolutions

Dans le premier cours, nous avons rapidement évoqué le modèle de données utilisé dans les bases de données relationnelles. Pour être franc, cette vision des bases de données est basée sur la version SQL-92 du standard. Depuis le début des années 2000, plusieurs évolutions de ce dernier ont enrichis les possibilités offertes par SQL. 

Si vous êtes allé explorer le site "_use-the-index-Luke_" (cf. moodle), vous êtes peut-être tombé sur une vidéo retraçant les évolutions de ce standard. 
En résumé, à l'origine le SQL a été conçu pour manipuler et gérer des ensembles de données pour pouvoir ensuite les traiter efficacement avec d'autres langages/outils informatique.

Les évolutions qui se sont succédées depuis ont permis d'enrichir la sémantique des informations manipulées. Aujourd'hui, il est toujours possible de manipuler des données "simple", mais il est également possible de gérer des informations de types complexes (Tableaux, Enregistrement...). 

## Le format JSON

Le format d'échange de données **JSON** (**J**ava**S**cript **O**bject **N**otation) est un mode de représentation sans schéma, en texte clair, de données structurées basées sur des paires _nom/valeur_ et des listes ordonnées.
C'est un format de données générique avec un nombre minimal de valeurs types : chaînes, nombres, booléens, listes, objets et null.

E.G.

    {
    “firstName”: “Morgan”,
    “lastName”: “Freeman”,
    “filmCount”: 4,
    “isActor”: true,
    “films”: [
                {
                    “title”: “Seven”,
                    “type”: “Thriller”
                },
                {
                    “title”: “Les évadés”,
                    “type”: “Policier”
                },
                {
                    “title”: “Amistad”,
                    “type”: “Historique”
                },
                {
                    “title”: “Oblivion”,
                    “type”: “Science Fiction”
                }
            ]
}

Ce petit exemple montre à quoi ressemble un fichier _JSON_. C'est un format d'échange de donnée simple dans lequel on retrouve, dans cet exemple, une petite partie de la filmographie de l'acteur Morgan Freeman (seulement 4 films dans cet exemple).

Puisque le format JSON est relativement simple à construire il est devenu, si ce n'est la norme, au moins un format d'échange de données privilégié entre différents langages informatique. De fait, c'est aujourd'hui le format de presque tous les services Web accessibles au public, et il est également utilisé fréquemment pour les services Web privés.  

La popularité de JSON tient également au fait que ce format est pris en charge nativement par de nombreuses bases de données. La norme SQL apporte ainsi, elle aussi, un support natif de JSON via des opérateurs dédiés. Attention toutefois, une règle pertinente consiste à utiliser ce formalisme aussi peu que possible dans le cadre de systèmes de bases de données relationnelles. En effet, ces dernières ont été optimisées pour des données structurées qui correspondent à un schéma bien particulier (ce qui n'est pas le cas avec JSON).

On constate que le résultat ne correspond pas à ce que l'on pourrait attendre...

Plusieurs options s'offrent à vous pour lire ce type de fichier, lesquelles ?

  * Vous pouvez écrire du code Python/Pandas pour re-traduire ce fichier dans un format plus simple et générer la Dataframe.
  * Vous pouvez tenter d'utiliser certaines subtilités offertent par les Pandas pour vous simplifier la vie.

```python
with open("./ExempleComplex.json",  'r') as f:
    data = json.loads(f.read())
```

On commence par lire le fichier avec une méthode offerte par JSON.

```python
data
```

Normalisation des données.

```python
DataSet_JSON_MN = pd.json_normalize(data, 
                                    record_path =['Results'], 
                                    meta =['original_number_of_clusters','Scaler','family_min_samples_percentage'], 
                                    meta_prefix='config_params_', 
                                    record_prefix='dbinfo_')
```

La methode utilisée pour la normalisation d'un fichier JSON offre plusieurs options, voici celles retenues:
record_path, record_prefix, meta et meta_prefix.

Explications:

  + **record_path**

C'est le "chemin" vers les éléments imbriqués (la structure) que l'on souhaite faire ressortir, comme pour les étaler devant soit.

  + **record_prefix**

Modification du préfixe par défaut en _dbinfo__.

  + **meta**

Liste des métadonnées que l'on souhaite stocker avec le record_path dans la dataframe (i.e. Scaler, original_number_of_clusters et family_min_samples_percentage).

  + **meta_prefix**

Même idée que précédement: modifier le préfix par défaut pour ces métadonnées. 

Sauvegarde au format CSV

```python
DataSet_JSON_MN.to_csv('DataSet_M_Norm.csv', index=False)
```
